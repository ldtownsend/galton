```python
import pyarrow.dataset as ds
import pyarrow as pa
import pandas as pd

dataset = ds.dataset(accuweather_path_1, format="parquet")
scanner = dataset.scanner()

dedup_cols = [c for c in dataset.schema.names if c != "model_timestamp"]

seen = set()
dedup_batches = []

for batch in scanner.to_batches():
    df = batch.to_pandas()
    # sort so earliest timestamp is first
    df = df.sort_values("model_timestamp")
    # define keys on dedup columns
    keys = list(zip(*(df[col].astype(str) for col in dedup_cols)))
    mask = []
    for k in keys:
        if k in seen:
            mask.append(False)
        else:
            seen.add(k)
            mask.append(True)
    df = df[mask]
    dedup_batches.append(pa.RecordBatch.from_pandas(df))

result_table = pa.Table.from_batches(dedup_batches)
```

```python
con = duckdb.connect()

con.execute("""
    CREATE OR REPLACE VIEW forecasts AS
    SELECT *
    FROM read_parquet('data/local_data/forecasts/**/*.parquet');
""")

con.execute("""
    CREATE OR REPLACE VIEW nws AS
    SELECT *
    FROM read_parquet('data/local_data/nws_observations/**/*.parquet');
""")
```

## Merging on Views w/ SQL

```python
con.execute(
    """
    SELECT forecasts.datetime_utc, forecast_temperature, model_name, nws.obs_temperature
    FROM forecasts
    JOIN nws 
        ON forecasts.city = nws.city 
        AND forecasts.datetime_utc = nws.datetime_utc
    WHERE forecasts.city = 'Austin' 
        AND forecasts.year_month = '2025-10';
    """
)

df_austin = con.fetch_df()
```
